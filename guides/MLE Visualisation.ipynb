# Maximum Likelihood Estimation

The core capability of the surpyval package is the ability to esimtate parameters of statistical distributions. the most powerful of the methods to estimate parameters is the Maximum Likelihood Estimation (MLE).

But what is it?

MLE, when first written down looks something like this:

$$e^x=\sum_{i=0}^\infty \frac{1}{i!}x^i$$

This is hardly an intuitive introduction. So what does this try to achieve?

Essentially, the MLE tried to find the most likely parameters to explain the data. In this case, the likelihood is maximised, where the likelihood can be thought of as the height of the line.

```python
import surpyval as surv
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist = surv.parametric.Weibull
```

```python
alpha = 10
beta = 5
samples = 100
t = dist.random(samples, alpha, beta)
c = np.zeros_like(t)
x = np.linspace(0, 20, 100)
```

To help us visualise maxmimum likelihood estimation for failures and left and right censoring.

```python
c[t > 12] = 1
t[t > 12] = 12
```

```python
c[t < 8] = -1
t[t < 8] = 8
```

```python
alpha = 5
beta = 3
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
f.set_figheight(4)
f.set_figwidth(15)
f.suptitle('LogLike: ' + str(-123.4) + ', alpha: ' + str(alpha) + ', beta: ' + str(beta), fontsize='x-large')
ax1.plot(x, dist.ff(x, alpha, beta))
for dt, cc in zip(t, c):
    if cc == -1:
        ax1.plot((dt, dt), (0, dist.ff(dt, alpha, beta)), 'k-')
    ax1.set_title(str(sum(c == -1)) + ' points')
ax2.plot(x, dist.df(x, alpha, beta))
for dt, cc in zip(t, c):
    if cc == 0:
        ax2.plot((dt, dt), (0, dist.df(dt, alpha, beta)), 'k-')
    ax2.set_title(str(sum(c == 0)) + ' points')
ax3.plot(x, dist.sf(x, alpha, beta))
for dt, cc in zip(t, c):
    if cc == 1:
        ax3.plot((dt, dt), (0, dist.sf(dt, alpha, beta)), 'k-')
    ax3.set_title(str(sum(c == 1)) + ' points')
f.subplots_adjust(top=0.85)
```
